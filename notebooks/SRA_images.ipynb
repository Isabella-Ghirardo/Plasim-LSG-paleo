```python
import xarray as xr
import matplotlib.pyplot as plt
import numpy as np

ds = xr.open_dataset('0129_cdo.nc')

data = ds['sg']/9.81

plt.figure(figsize=(12, 6))

data.plot(
    cmap='terrain', 
    cbar_kwargs={'label': 'Elevation [m]'}
)

plt.title("Eocene Orography Map (from 0129_cdo.nc)")

plt.savefig('0129_orography.png', dpi=300, bbox_inches='tight')

plt.show()
```

```python
import xarray as xr
import matplotlib.pyplot as plt

ds = xr.open_dataset('0172_cdo.nc')
data = ds['lsm']

plt.figure(figsize=(12, 6))
# 'bone' is good for binary: Black (0) for ocean, White (1) for land
data.plot(cmap='viridis', cbar_kwargs={'label': '0=Ocean, 1=Land'}) 

plt.title("Land-Sea Mask (0172)")
plt.savefig('0172_lsm.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
import xarray as xr
import matplotlib.pyplot as plt

ds = xr.open_dataset('0212.nc')
# Check variable name: usually 'vegf' or 'var212'
data = ds['forest_ratio']

plt.figure(figsize=(12, 6))
data.plot(cmap='Greens', vmin=0, vmax=1, cbar_kwargs={'label': 'Fraction'})

plt.title("Forest Ratio (0212)")
plt.savefig('0212_forest_ratio.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
import xarray as xr
import matplotlib.pyplot as plt

ds = xr.open_dataset('0174_cdo.nc')
data = ds['albedo']

plt.figure(figsize=(12, 6))
data.plot(cmap='gist_earth', vmin=0.0, vmax=0.5, cbar_kwargs={'label': 'Albedo'})

plt.title("Surface Background Albedo (0174)")
plt.savefig('0174_albedo.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
import xarray as xr
import matplotlib.pyplot as plt

ds = xr.open_dataset('0173.nc')
data = ds['z0']

plt.figure(figsize=(12, 6))
# 'copper_r' makes high roughness dark, low roughness light
data.plot(cmap='copper_r', vmin=0, vmax=2.5, cbar_kwargs={'label': 'Roughness [m]'})

plt.title("Roughness Length (0173)")
plt.savefig('0173_roughness.png', dpi=300, bbox_inches='tight')
plt.show()
```
